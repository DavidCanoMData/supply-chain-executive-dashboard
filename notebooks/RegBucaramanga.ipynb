```python
import pandas as pd
import numpy as np

# Cargar dataset
df = pd.read_csv(r"C:\Users\cano9\Downloads\aldia_operaciones_Q1_2026.csv")

# Primer vistazo
print(f"Filas: {df.shape[0]:,}  |  Columnas: {df.shape[1]}")
print("\nTipos de datos:")
print(df.dtypes)
print("\nPrimeras 3 filas:")
df.head(3)
```

```python
df = df.rename(columns={
    "remesa_id":                "ID_Remesa",
    "fecha":                    "Fecha",
    "año":                      "Año",
    "trimestre":                "Trimestre",
    "mes":                      "Mes",
    "mes_num":                  "Mes_Numero",
    "semana_año":               "Semana",
    "dia_semana":               "Dia_Semana",
    "regional":                 "Regional",
    "tipo_operacion":           "Tipo_Operacion",
    "conductor_id":             "ID_Conductor",
    "conductor_nombre":         "Conductor",
    "zona_entrega":             "Zona_Entrega",
    "vehiculo":                 "Vehiculo",
    "tipo_mercancia":           "Tipo_Mercancia",
    "peso_kg":                  "Peso_KG",
    "unidades":                 "Unidades",
    "hora_salida_plataforma":   "Hora_Salida",
    "hora_entrega":             "Hora_Entrega",
    "tiempo_gestion_min":       "Tiempo_Gestion_Min",
    "estado_entrega":           "Estado_Entrega",
    "entregado":                "Entregado",
    "novedad":                  "Novedad"
})

print("Columnas renombradas:")
print(df.columns.tolist())
```

```python
dias_es = {
    "Monday":"Lunes", "Tuesday":"Martes", "Wednesday":"Miércoles",
    "Thursday":"Jueves", "Friday":"Viernes", "Saturday":"Sábado", "Sunday":"Domingo"
}
meses_es = {
    "January":"Enero", "February":"Febrero", "March":"Marzo",
    "April":"Abril", "May":"Mayo", "June":"Junio",
    "July":"Julio", "August":"Agosto", "September":"Septiembre",
    "October":"Octubre", "November":"Noviembre", "December":"Diciembre"
}

df["Dia_Semana"] = df["Dia_Semana"].map(dias_es)
df["Mes"]        = df["Mes"].map(meses_es)

print("Días únicos:", df["Dia_Semana"].unique())
print("Meses únicos:", df["Mes"].unique())
```

```python
# Fecha como tipo fecha (no texto)
df["Fecha"] = pd.to_datetime(df["Fecha"])

# Columna: ¿fue entregado exitosamente? (Sí / No)
df["Entrega_Exitosa"] = df["Entregado"].map({1: "Sí", 0: "No"})

# Columna: categoría de peso
def categorizar_peso(kg):
    if kg < 5:       return "Liviano (<5 kg)"
    elif kg < 50:    return "Mediano (5-50 kg)"
    elif kg < 200:   return "Pesado (50-200 kg)"
    else:            return "Carga masiva (>200 kg)"

df["Categoria_Peso"] = df["Peso_KG"].apply(categorizar_peso)

# Columna: categoría de tiempo de gestión
def cat_tiempo(min):
    if min <= 30:    return "Rápido (≤30 min)"
    elif min <= 90:  return "Normal (31-90 min)"
    elif min <= 180: return "Demorado (91-180 min)"
    else:            return "Crítico (>180 min)"

df["Categoria_Tiempo"] = df["Tiempo_Gestion_Min"].apply(cat_tiempo)

print("Nuevas columnas creadas:")
print(df[["Entrega_Exitosa","Categoria_Peso","Categoria_Tiempo"]].value_counts().head(10))
```

```python
print("Valores nulos por columna:")
print(df.isnull().sum())

# La columna Novedad tiene nulos válidos (remesas sin novedad = cadena vacía)
df["Novedad"] = df["Novedad"].fillna("Sin novedad")

print("\n¿Duplicados?", df.duplicated().sum())
print("\nDataset limpio listo. Forma final:", df.shape)
```

```python
df.to_csv("aldia_Q1_2026_limpio.csv", index=False, encoding="utf-8-sig")
print("✅ Archivo exportado: aldia_Q1_2026_limpio.csv")
print(f"   {df.shape[0]:,} filas  |  {df.shape[1]} columnas")
```
